```python
import pandas as pd
import requests
import json
from pprint import pprint
import pandas as pd
```

```python
province_dict = {
    'City': None,
     'Time Zone': None,
     'Country': None,
     'Region': None,
     'City': None,
     'Longitude': None,
     'Latitude': None,
     'Population': None,
}
```

```python
import csv
csv_file = r"mycsv.csv"

data_all_province = pd.DataFrame();
province_list = ['Chiang_Rai','Chiang_Mai']
for province in province_list:
    p = province_dict.copy()
    url2 = f'http://berkeleyearth.lbl.gov/air-quality/maps/cities/Thailand/{province}/{province}.txt' #แสดงค่าที่ได้จากจุดตรวจวัดล่าสุด
    params = { }
    response = requests.request('GET', url2 ,params=params)
    print(response.text)
    
    in_txt = csv.reader(response.text, delimiter = '\t')
    print(in_txt)
    data=  response.text.json()  #to extract the detail from response
    
    hour_data = pd.DataFrame(data['hour_data']);
    p['City'] = data['City']
    p['Time Zone'] = data['Time Zone']
    p['Country'] = data['Country'] 
    p['Region'] = data['Region']
    p['City'] = data['City (ASCII)']
    p['Longitude'] = data['Longitude']
    p['Latitude'] = data['Latitude']
    p['Population'] = data['Population']
   
    #del data['day_data'];
    print(p)
    hour_data.to_csv('data\\'+province + '_data.csv')
    data_all_province = data_all_province.append(p, ignore_index=True)
data_all_province.to_csv('data_all_province.csv')
```
